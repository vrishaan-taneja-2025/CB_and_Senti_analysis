```python
import pandas as pd
import random
import re

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("data/test_chatbot_intents.csv")

print("Columns:", df.columns)
print("Total samples:", len(df))

df.head()
```

```python
def clean_text(text):
    text = str(text).lower()

    text = text.replace("nlp", "natural language processing")
    text = text.replace("ai", "artificial intelligence")

    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
train_df, test_df = train_test_split(
    df,
    test_size=0.6,
    stratify=df["intent"],
    random_state=42
)

print("Train:", len(train_df))
print("Test:", len(test_df))
```

```python
prefixes = ["", "can you", "please", "i want to know", "tell me"]
suffixes = ["", "please", "in simple words", "quickly"]

synonyms = {
    "what is": ["define", "explain", "tell me about"],
    "nlp": ["natural language processing"],
    "ai":["artificial intelligence"]
}
```

```python
def augment_text(text):
    text = str(text).lower()

    for key, values in synonyms.items():
        if key in text:
            text = text.replace(key, random.choice(values))

    new_text = f"{random.choice(prefixes)} {text} {random.choice(suffixes)}"
    return " ".join(new_text.split())
```

```python
augmented_data = []

for _, row in train_df.iterrows():
    augmented_data.append(row.to_dict())  # original

    for _ in range(3):  # augmentation factor
        new_text = augment_text(row["utterance"])
        augmented_data.append({
            "utterance": new_text,
            "intent": row["intent"]
        })

train_aug_df = pd.DataFrame(augmented_data)

print("Augmented train size:", len(train_aug_df))
```

```python
train_aug_df = train_aug_df.drop_duplicates(subset=["utterance"])

print("After removing duplicates:", len(train_aug_df))
```

```python
train_aug_df["clean"] = train_aug_df["utterance"].apply(clean_text)
test_df["clean"] = test_df["utterance"].apply(clean_text)
```

```python
print("Train class distribution:\n")
print(train_aug_df["intent"].value_counts())
```

```python
print("Test class distribution:\n")
print(test_df["intent"].value_counts())
```

```python
train_set = set(train_aug_df["utterance"])
test_set = set(test_df["utterance"])

overlap = train_set.intersection(test_set)

print("Overlap:", len(overlap))
```

```python
vectorizer = TfidfVectorizer(
    stop_words="english",
    ngram_range=(1, 2)
)

X_train = vectorizer.fit_transform(train_aug_df["clean"])
X_test = vectorizer.transform(test_df["clean"])
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, train_aug_df["intent"])

print("Model trained")
```

```python
preds = model.predict(X_test)

accuracy = accuracy_score(test_df["intent"], preds)

print("Accuracy:", accuracy)
print("\nClassification Report:\n")
print(classification_report(test_df["intent"], preds))
```

```python
import matplotlib.pyplot as plt

cm = confusion_matrix(test_df["intent"], preds)
```

```python
labels = model.classes_

plt.figure(figsize=(6,5))
plt.imshow(cm)

plt.xticks(range(len(labels)), labels, rotation=45)
plt.yticks(range(len(labels)), labels)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.colorbar()
plt.show()
```

```python
import numpy as np

cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

plt.imshow(cm_norm)
plt.title("Normalized Confusion Matrix")
plt.colorbar()
plt.show()
```

```python
for i in range(len(labels)):
    for j in range(len(labels)):
        if i != j and cm[i][j] > 0:
            print(f"{labels[i]} → {labels[j]}: {cm[i][j]}")
```

```python
responses = {
    "greeting": [
        "Hello! How can I help you?",
        "Hi there! Ask me about NLP or AI."
    ],
    "nlp_definition": [
        "NLP helps computers understand human language.",
        "It is a field of AI focused on text processing."
    ],
    "tokenization": [
        "Tokenization splits text into smaller units.",
    ],
    "stemming": [
        "Stemming reduces words to their root form.",
    ],
    "lemmatization": [
        "Lemmatization converts words to dictionary form.",
    ],
    "ai_applications": [
        "AI is used in healthcare, finance, and automation.",
    ],
    "goodbye": [
        "Goodbye! Have a nice day."
    ],
    "fallback": [
        "Sorry, I didn’t understand that."
    ]
}
```

```python
def predict_intent(text):
    text_clean = clean_text(text)
    vec = vectorizer.transform([text_clean])

    probs = model.predict_proba(vec)[0]
    idx = probs.argmax()

    intent = model.classes_[idx]
    confidence = probs[idx]
    print(confidence)
    if confidence < 0.25:
        return "fallback"

    return intent
```

```python
def chatbot(text):
    intent = predict_intent(text)
    return random.choice(responses[intent])
```
